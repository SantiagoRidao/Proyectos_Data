```python
import pandas as pd
delitos_2016 = pd.read_csv('/content/drive/MyDrive/ML/ArgPRograma/Datasets/delitos_2016.csv')
delitos_2016.head()
```

```python
delitos_2017 = pd.read_csv('/content/drive/MyDrive/ML/ArgPRograma/Datasets/delitos_2017.csv')
delitos_2017.head()
```

```python
delitos_2018= pd.read_csv('/content/drive/MyDrive/ML/ArgPRograma/Datasets/delitos_2018.csv')
delitos_2018.head()
```

```python
delitos = pd.concat([delitos_2016, delitos_2017, delitos_2018], ignore_index=True)
delitos.head()
```

```python
delitos.isnull().sum()
```

```python
# Separar la columna 'fecha' en año, mes y día
delitos['anio'] = pd.to_datetime(delitos['fecha']).dt.year
delitos['mes'] = pd.to_datetime(delitos['fecha']).dt.month
delitos['dia'] = pd.to_datetime(delitos['fecha']).dt.day

# Convertir la columna 'fecha' a objetos de fecha si no está en ese formato
delitos['fecha'] = pd.to_datetime(delitos['fecha'])


# Crear una nueva columna 'dia_semana' con el nombre del día de la semana
delitos['dia_semana'] = delitos['fecha'].dt.day_name()

def traducir_dias_semana(dataframe, columna_fecha):
    # Diccionario de traducción de días de la semana
    traducciones = {
        'Monday': 'Lunes',
        'Tuesday': 'Martes',
        'Wednesday': 'Miércoles',
        'Thursday': 'Jueves',
        'Friday': 'Viernes',
        'Saturday': 'Sábado',
        'Sunday': 'Domingo'
    }

    # Convertir la columna de fecha a objetos de fecha si no está en ese formato
    dataframe[columna_fecha] = pd.to_datetime(dataframe[columna_fecha])

    # Crear una nueva columna 'dia_semana' con el nombre del día de la semana en español
    dataframe['dia_semana'] = dataframe[columna_fecha].dt.day_name()

    # Reemplazar los nombres de los días de la semana en inglés por los equivalentes en español
    dataframe['dia_semana'] = dataframe['dia_semana'].replace(traducciones)

# Llamamos a la función para traducir y agregar la columna 'dia_semana' en español
traducir_dias_semana(delitos, 'fecha')

# Mostrar el DataFrame resultante
# Mostrar el DataFrame resultante

delitos.drop(columns=['fecha'], inplace=True)
```

```python
# Crear un gráfico de barras
plt.figure(figsize=(6, 6))
delitos_cleaned['dia_semana'].value_counts().plot(kind='bar')

# Configurar etiquetas y título
plt.xlabel('Día de la Semana')
plt.ylabel('Cantidad de Delitos')
plt.title('Cantidad de Delitos por Día de la Semana')

# Mostrar el gráfico
plt.show()
```

```python
# Calcular la distribución de delitos por día de la semana
distribucion_dias = delitos_cleaned['dia_semana'].value_counts()

# Crear un gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(distribucion_dias, labels=distribucion_dias.index, autopct='%1.1f%%', startangle=140)

# Configurar el título
plt.title('Distribución de Delitos por Día de la Semana')

# Mostrar el gráfico de pastel
plt.show()
```

La Figura  muestra que la cantidad de delitos disminuye
en los dıas Domingos.Nos indica que las actividades delictivas esta
relacionadas con el nivel de actividad de la poblacion lo cual
tiene bastante sentido, ya que mientras mas movimiento haya,
mas alta es la probabilidad de que ocurran accidentes y mas
gente ocuparan las calles, dando lugar a los asaltos y hurtos.

```python
# Calcular la distribución de delitos por barrio
distribucion_barrios = delitos_cleaned['barrio'].value_counts()

# Tomar los 20 barrios con más delitos
# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
distribucion_barrios.plot(kind='bar', color='skyblue')

# Configurar etiquetas y título
plt.xlabel('Barrio')
plt.ylabel('Cantidad de Delitos')
plt.title('Top 10 Barrios con Más Delitos en Buenos Aires')

# Mostrar el gráfico
plt.show()
```

Se puede observar sin lugar a dudas, en contraste con los
graficos  que el barrio Palermos presenta la mayor
cantidad de delitos, seguido por Balvanera y Flores.

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado 'delitos_cleaned' con una columna 'barrio'
# y las coordenadas 'lat' y 'long'

# Cargar los límites geográficos de los barrios de la Ciudad de Buenos Aires
barrios_geo = gpd.read_file('https://cdn.buenosaires.gob.ar/datosabiertos/datasets/barrios/barrios.geojson')

# Crear un mapa de Buenos Aires
fig, ax = plt.subplots(figsize=(10, 12))
barrios_geo.plot(ax=ax, color='lightgray', edgecolor='black')

# Crear un mapa de calor hexagonal utilizando hexbin
hb = ax.hexbin(
    delitos_cleaned['long'],
    delitos_cleaned['lat'],
    gridsize=50,  # Tamaño de la malla hexagonal
    bins='log',  # Escala logarítmica para la cantidad de delitos
    cmap='YlOrRd',  # Paleta de colores
    alpha=0.7,  # Transparencia
    mincnt=1,  # Mínimo número de delitos por hexágono
)

# Agregar barra de color
cb = plt.colorbar(hb, ax=ax)
cb.set_label('Cantidad de Delitos (escala logarítmica)')

# Configurar etiquetas y título
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')
ax.set_title('Mapa de Calor Hexagonal de Delitos por Barrio en Buenos Aires')

# Mostrar el mapa
plt.show()

```

```python
import seaborn as sns

# Distribución de delitos por tipo de delito
plt.figure(figsize=(10, 6))
sns.countplot(data=delitos_cleaned, y='tipo_delito')
plt.title('Distribución de Delitos por Tipo de Delito')
plt.xlabel('Cantidad de Delitos')
plt.ylabel('Tipo de Delito')
plt.show()
```

```python
# Matriz de correlación entre variables numéricas
corr_matrix = delitos_cleaned.corr()

# Visualizar la matriz de correlación con un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlación')
plt.show()
```
